# 2. Leitura

Funções

```python
def normaliza_habilidades(txt):
    txt_novo = txt.replace('[','').replace(']','').replace("'",'').strip()
    
    return txt_novo
    
```

Diretório

```python
dir_arq = '../1.Bases'
```

Dados dos Pokemons

```python
# gosto de fazer a leitura dessa forma para caso entre mais arquivos semelhantes, já concatene num só dataframe. 
#Também gosto de manter um "raw" dataframe para caso dê algum problema no código no futuro, eu não precise fazer a leitura novamente.
```

```python
r_df_pokemons = pd.DataFrame()

for arq in listdir(dir_arq):
    aux = pd.read_csv(f'{dir_arq}/{arq}', sep = ',', encoding='utf-8')
    
    r_df_pokemons = pd.concat([r_df_pokemons, aux])
```

# 3. Processamento

```python
#aqui um exemplo de conseguir retornar ao dataframe original sem precisar refazer a leitura
df_pokemons = r_df_pokemons.copy()
```

```python
# Tô pensando no dashboard, então vou separar o dataframe em algumas tabelas menores que relacionarei no power bi
```

```python
# criando uma tabela dimensão com o nome dos pokemons

ls_cols_info = ['name','attack','classfication', 'defense', 'height_m', 'hp','japanese_name', 'percentage_male', 'pokedex_number', 'sp_attack', 'sp_defense', 'speed', 'type1', 'type2', 'weight_kg','generation', 'is_legendary']

dim_infos_pokemon = df_pokemons[ls_cols_info].drop_duplicates()
```

```python
# separando as habilidades em um dataframe
df_habilidades = df_pokemons[['name', 'abilities']].copy()

#normalizando os nomes que estavam diferentes
df_habilidades['abilities'] = df_habilidades.abilities.apply(normaliza_habilidades)
```

```python
# separando as colunas que dizem os pontos de batalha de cada pokemon (todas que começam com against)
ls_cols_batalha = [coluna for coluna in df_pokemons.columns if coluna.startswith('against')]

df_batalhas = df_pokemons[['name'] + ls_cols_batalha]

# transformando as colunas em linhas porque acho que assim fica mais fácil de manipular no power bi
df_batalhas = df_batalhas.melt(id_vars = 'name', value_vars = ls_cols_batalha, var_name = 'battle', value_name = 'points')
```

# 4. Exportação

```python
writer = pd.ExcelWriter('../2.Saídas/desafio_machiron_gabriellamaia.xlsx', engine='xlsxwriter')
```

```python
dim_infos_pokemon.to_excel(writer, index=False, sheet_name='dim_nomes_pokemon')
df_habilidades.to_excel(writer, index=False, sheet_name='habilidades')
df_batalhas.to_excel(writer, index=False, sheet_name='batalhas')
```

```python
#bibliotecas estatísticas
import statistics
#import math
#from scipy import stats
from collections import Counter
import matplotlib.pyplot as plt
```

```python
# vou começar fazendo as estatísticas descritivias: moda, média e mediana. Escolhi analisar a HP por tipo de pokemon.
```

```python
#a moda nos diz o valor que mais se repete, portanto eu consigo saber qual o valor de HP mais comum em cada tipo de pokemon
#e também sei que em geral o tipo fairy tem pokémons com hps mais altos

ls_tipos = []
ls_modas = []
for tipo in dim_infos_pokemon.type1.unique():
    df_aux = dim_infos_pokemon[dim_infos_pokemon['type1'] == tipo].copy()
    moda = statistics.mode(list(df_aux.hp))
    ls_tipos = ls_tipos + [tipo]
    ls_modas = ls_modas + [moda]
    
resumo_moda = pd.DataFrame((zip(ls_tipos, ls_modas)), columns = ['tipo', 'moda'])
resumo_moda
```

```python
#a média nos diz o valor "médio" de hp de cada tipo. 
#Não gosto de me basear pela média porque imagino que cada tipo tenha valores máximos e mínimos de hp bem distintos, o que pode enviesar essa média
#a média nos diz que em geral o tipo dragon tem HP mais alto

ls_tipos = []
ls_medias = []
for tipo in dim_infos_pokemon.type1.unique():
    df_aux = dim_infos_pokemon[dim_infos_pokemon['type1'] == tipo].copy()
    media = round(statistics.mean(list(df_aux.hp)),0)
    ls_tipos = ls_tipos + [tipo]
    ls_medias = ls_medias + [media]
    
resumo_media = pd.DataFrame((zip(ls_tipos, ls_medias)), columns = ['tipo', 'media'])
resumo_media
```

```python
#a mediana nos diz o valor central de cada amostra. Dá pra perceber que os valores das medianas variam menos entre os tipos que o da média e moda
#pela análise da mediana eu escolheria o tipo flying 

ls_tipos = []
ls_medianas = []
for tipo in dim_infos_pokemon.type1.unique():
    df_aux = dim_infos_pokemon[dim_infos_pokemon['type1'] == tipo].copy()
    mediana = round(statistics.median(list(df_aux.hp)),0)
    ls_tipos = ls_tipos + [tipo]
    ls_medianas = ls_medianas + [mediana]
    
resumo_mediana = pd.DataFrame((zip(ls_tipos, ls_medianas)), columns = ['tipo', 'mediana'])
resumo_mediana
```

```python
# agora eu quero testar se existe diferença entre os gêneros no pokémon em relação às habilidades
```

```python
batalha_genero = dim_infos_pokemon[['name','percentage_male','attack','defense']].copy()
```

```python
batalha_genero['genero'] = batalha_genero.apply(lambda x: 'macho' if x['percentage_male']>50 else 'femea', axis = 1)
```

```python
# pela média as fêmeas são mais fracas em ataque mas melhores em defesa

batalha_genero.groupby(['genero']).agg({'attack':'mean', 'defense':'mean'}).reset_index()
```

```python
# quero saber a distribuição dos pokémons por peso, somente do tipo fairy, que são meus favoritos 

dim_infos_pokemon.query("type1=='fairy'").groupby(['name']).agg({'weight_kg':'sum'}).reset_index().sort_values(by='weight_kg', ascending=False)
```

```python
# quero saber o tipo de pokémon com o ataque mais forte

# é o tipo bug

dim_infos_pokemon.groupby(['type1']).agg({'attack':'max'}).reset_index().sort_values(by='attack', ascending=False)
```

```python
# quero saber o tipo de pokémon com a defesa mais forte

# é o tipo bug também, logo eu escolheria um bug como inicial

dim_infos_pokemon.groupby(['type1']).agg({'defense':'max'}).reset_index().sort_values(by='defense', ascending=False)
```
